```python
df_model = df[['Molecular_weight MW (g/mol)', 'pKa', 'Log P','Log S','Polar Surface Area (Å2)','Polarizability (Å3)','HOMO (eV)','LUMO (eV)','Electronegativity (eV)',' ΔN_Fe ','IE EXP (%)']] = df[['Molecular_weight MW (g/mol)', 'pKa', 'Log P','Log S','Polar Surface Area (Å2)','Polarizability (Å3)','HOMO (eV)','LUMO (eV)','Electronegativity (eV)',' ΔN_Fe ','IE EXP (%)']]
```

```python
df_model["Molecular_weight MW (g/mol)"] = df_model["Molecular_weight MW (g/mol)"].astype(float)
df_model["Log P"] = df_model["Log P"].astype(float)
df_model["Log S"] = df_model["Log S"].astype(float)
df_model["Polar Surface Area (Å2)"] = df_model["Polar Surface Area (Å2)"].astype(float)
df_model["Polarizability (Å3)"] = df_model["Polarizability (Å3)"].astype(float)
```

```python
models = {
    #linier
    # "Linear Regression": LinearRegression(),
    # "Ridge" : Ridge(),
    # "Lasso" : Lasso(),
    # "Elastic Net" : ElasticNet(),
    #nonlinier
    # "XGB Regressor" : XGBRegressor(),
    # "CatBoost Regressor" : CatBoostRegressor(),
    # "Light GBM Regressor" : LGBMRegressor(),
    "Gradient Boosting Regressor" : GradientBoostingRegressor(),
    "AdaBoost Regressor" : AdaBoostRegressor(),
    # "NuSVR" : NuSVR(),
    # "Decision Tree Regressor": DecisionTreeRegressor(),
    # "SVR": SVR(),
    "Random Forest Regressor":RandomForestRegressor(),
}
```

```python
# Variasi metode normalisasi
scalers = [RobustScaler(), StandardScaler(), MinMaxScaler(), "None"]

# Variasi pembagian data
test_sizes = [0.2, 0.3, 0.4]

list_test_size = []
list_scaler = []
list_model_name = []
list_rmse = []
list_r2 =[]

# Loop untuk kombinasi
for model_name, model in models.items():
    for scaler in scalers:
        for test_size in test_sizes:
            # Splitting X dan y
            X = df_model[['Molecular_weight MW (g/mol)', 'pKa', 'Log P', 'Log S', 'Polar Surface Area (Å2)', 'Polarizability (Å3)', 'HOMO (eV)', 'LUMO (eV)', 'Electronegativity (eV)', ' ΔN_Fe ']]
            y = df_model['IE EXP (%)']

            # Bagi data
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

            # Normalisasi jika ada scaler
            if scaler != "None":
                X_train_scaled = scaler.fit_transform(X_train)
                X_test_scaled = scaler.transform(X_test)
                # y dinormalisasi
#                 y_train_scaled = scaler.fit_transform(y_train.values.reshape(-1, 1)).flatten()
#                 y_test_scaled = scaler.transform(y_test.values.reshape(-1, 1)).flatten()

            else:
                X_train_scaled = X_train
                X_test_scaled = X_test
                # y dinormalisasi
#                 y_train_scaled = y_train
#                 y_test_scaled = y_test

            # Latih model
            model.fit(X_train_scaled, y_train)

            # Prediksi dan evaluasi
            y_pred = model.predict(X_test_scaled)
            mse = mean_squared_error(y_test, y_pred)
            rmse = np.sqrt(mse)
            r2 = r2_score(y_test, y_pred)

            print(f"{model_name} | Scaler: {scaler} | Test Size: {test_size} | RMSE: {rmse} | R2 : {r2}")
            
            custom_path = f"{model_name}_{test_size}_{scaler}"
            
#             # save list akurasi dalam bentuk xls
#             list_test_size.append(test_size)
#             list_scaler.append(scaler)
#             list_model_name.append(model_name)
#             list_rmse.append(rmse)
#             list_r2.append(r2)
            
#             header = ["Test_Size", "Scaler", "Model", "RMSE", "R2"]
#             dataframe_model = pd.DataFrame(list(zip(list_test_size,list_scaler,list_model_name,list_rmse,list_r2)),columns=header)
#             excel_file = "HOCV_Testing_Result.xlsx"
#             dataframe_model.to_excel(excel_file,index=False)
            
            # Save the model using joblib
            joblib.dump(model, f'{custom_path}.joblib')

            # Save the model using pickle
            with open(f'{custom_path}.pkl', 'wb') as model_file:
                pickle.dump(model, model_file)
```

```python

# Definisikan jumlah lipatan (folds) yang diinginkan
num_folds_list = [3, 5, 10]

# Variasi metode normalisasi
scalers = [RobustScaler(), StandardScaler(), MinMaxScaler(), "None"]

list_num_folds = []
list_scaler = []
list_model_name = []
list_rmse = []
list_r2 =[]

# Loop untuk kombinasi
for model_name, model in models.items():
    for scaler in scalers:
        for num_folds in num_folds_list:
            # Splitting X dan y
            X = df_model[['Molecular_weight MW (g/mol)', 'pKa', 'Log P', 'Log S', 'Polar Surface Area (Å2)', 'Polarizability (Å3)', 'HOMO (eV)', 'LUMO (eV)', 'Electronegativity (eV)', ' ΔN_Fe ']]
            y = df_model['IE EXP (%)']

#             # Normalisasi jika ada scaler
#             if scaler != "None":
#                 X_scaled = scaler.fit_transform(X)
#             else:
#                 X_scaled = X
                
            # Normalisasi scaler x dan y
            if scaler != "None":
                X_scaled = scaler.fit_transform(X)
#                 y_scaled = scaler.fit_transform(y.values.reshape(-1, 1)).flatten()
            else:
                X_scaled = X
                y_scaled = y

            # Latih model
            model.fit(X_scaled, y)

            kfold = KFold(n_splits=num_folds, shuffle=True, random_state=42)

            # Define scoring metrics
            scoring = {'r2': make_scorer(r2_score), 'rmse': make_scorer(mean_squared_error, squared=False)}

            # Lakukan validasi silang
            r2_scores = cross_val_score(model, X_scaled, y, cv=kfold, scoring=scoring['r2'])
            rmse_scores = cross_val_score(model, X_scaled, y, cv=kfold, scoring=scoring['rmse'])

#             # Hitung rata-rata MSE dan ubah ke positif
#             avg_mse = -np.mean(mse_scores)
#             avg_rmse = np.sqrt(avg_mse)
            rmse_scores = np.mean(rmse_scores)
            r2_scores = np.mean(r2_scores)
            print(f"{model_name} | Scaler: {scaler} | Folds: {num_folds} | RMSE: {rmse_scores} | R2-Scores : {r2_scores}")
            
            # # save list akurasi dalam bentuk xls
            # list_num_folds.append(num_folds)
            # list_scaler.append(scaler)
            # list_model_name.append(model_name)
            # list_rmse.append(rmse_scores)
            # list_r2.append(r2_scores)
            
            # header = ["Test_Size", "Scaler_KF", "Model", "RMSE", "R2"]
            # dataframe_model = pd.DataFrame(list(zip(list_num_folds,list_scaler,list_model_name,list_rmse,list_r2)),columns=header)
            # excel_file = "KFCV_Fit_Result.xlsx"
            # dataframe_model.to_excel(excel_file,index=False)

            # Simpan model menggunakan joblib
            custom_path = f"{model_name}_{num_folds}_{scaler}"
            joblib.dump(model, f'{custom_path}.joblib')

            # Simpan model menggunakan pickle
            with open(f'{custom_path}.pkl', 'wb') as model_file:
                pickle.dump(model, model_file)
```
