```python
df[['Molecular_weight MW (g/mol)', 'pKa', 'Log P','Log S','Polar Surface Area (Å2)','Polarizability (Å3)','HOMO (eV)','LUMO (eV)','Electronegativity (eV)',' ΔN_Fe ']] = df[['Molecular_weight MW (g/mol)', 'pKa', 'Log P','Log S','Polar Surface Area (Å2)','Polarizability (Å3)','HOMO (eV)','LUMO (eV)','Electronegativity (eV)',' ΔN_Fe ']].replace(',', '.', regex=True)
```

```python
df["Molecular_weight MW (g/mol)"] = df["Molecular_weight MW (g/mol)"].astype('float')
df["Log P"] = df["Log P"].astype('float')
df["Log S"] = df["Log S"].astype('float')
df["Polar Surface Area (Å2)"] = df["Polar Surface Area (Å2)"].astype('float')
df["Polarizability (Å3)"] = df["Polarizability (Å3)"].astype('float')
```

```python
df
```

```python
mean_molecular_weight = df["Molecular_weight MW (g/mol)"].mean()
mean_molecular_weight
```

```python
# Menghitung nilai rata-rata untuk setiap kolom
mean_values = df.mean()

# Mengisi nilai-nilai yang hilang dengan nilai rata-rata masing-masing kolom
df.fillna(mean_values, inplace=True)

# Menampilkan DataFrame yang telah diisi
df
```
